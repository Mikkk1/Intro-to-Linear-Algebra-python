## Accessing string characters

### Indexing and Slicing:
##### As, strings are sequences of characters, we can do indexing and slicing to grab sub-sections of the string.
##### In indexing a '[index_number]' is used after the 'string_variable_name', to get that specific index character.

*   Indexing starts at zero.
*   We can also do reverse indexing, e.g.,


##### -- String  : : H __ e __ l ___ l _ _ o

##### --- Index   :   [ 0] , [ 1] ,  [ 2] , [ 3] , [ 4]
##### Reverse :   [-5] , [-4] , [-3] , [-2] , [-1]


```python
name[0]
```

```python
name = "Sarim Zahid "
```

```python
print(
name[6],
name[-5] ,
name[-4],
name[-3],
name[-2],
name[-1])
```

### Slicing alows us to get a subsection of a string:

 It has the following Syntax:
>**[start: stop: step]**


* **Start:** is the number of index, for the slice start.  
* **Stop :**    is the number of index we want to go upto. (It will not be included in the slice.)
*   **Step :**  number of characters we want to skip, between each two characters.
#### Here, are a few examples:



```python
 print(name[0:7])
```

```python
 # name = Sarim Zahid
 print(name[3:8])
```

Jump of **characters**:

```python
print(name[0:11:3])
```

```python
st = "123456789"
```

```python
last_index = len(st) - 1
last_index
```

```python
print(st[1:8:2])
```

# 2: Lists and their basic functions:
* Used to represent group of Individual objects as single entity.
*  Represented by [obj1, obj2,....].
* It is a sequence of objects, so it is iteratable.
* Indexing and Slicing is allowed.
* Order is preserved.
* Duplicates are allowed.
* Heterogenous objects are allowed.
* List is dynamic, size can be increased or decreased.
* Mutable
