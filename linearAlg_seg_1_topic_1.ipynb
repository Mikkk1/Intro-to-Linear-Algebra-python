### What is Linear Algebra?

Sloving for unknowns in a system of linear equations.


```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining the time using linspace function in numpy; **np.linspace(start, end, n points)** :

```python
t = np.linspace(0,50,1000)
```

#### Distance (d) = st
Distance travelled by robber: d = 2.5t

```python
d_r = 2.5 * t
```

Distance travelled by Sherif: d = 3(t - 5)

```python
d_s = 3 * (t - 5)
```

#### Let's plot the output in graphical form using matplotlib:

```python
# Create a figure and an axis:
fig, ax = plt.subplots()

# Title and labels
plt.title("Bank robber caught")
plt.xlabel("Time (m)")
plt.ylabel("Distance (Km)")

# Set the x and y limits:
ax.set_xlim([0,40])
ax.set_ylim([0,100])

# Ploting the Distance covered by robber and sheriff:
ax.plot(t, d_r, c = 'green')
ax.plot(t, d_s, c = 'purple')

# As, we have solved it using linear algebra we know the answers. So,
# Plotting vertical line where robber was caught:
plt.axvline(x = 30, color = 'red', linestyle = "--")

# Plotting horizontal line where robber was caught:
plt.axhline(y = 75, color = 'red', linestyle = "--")
```
